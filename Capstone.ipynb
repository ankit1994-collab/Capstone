## EDA 

```python
import pandas as pd
import glob

schema = {
    'song_hotttnesss': float,
    'artist_familiarity': float,
    'artist_hotttnesss': float,
    'artist_id': str,
    'artist_latitude': float,
    'artist_location': str,
    'artist_longitude': float,
    'artist_name': str,
    'title': str,
    'danceability': float,
    'duration': float,
    'end_of_fade_in': float,
    'energy': float,
    'key': float,
    'key_confidence': float,
    'loudness': float,
    'mode': float,
    'mode_confidence': float,
    'start_of_fade_out': float,
    'tempo': float,
    'time_signature': float,
    'time_signature_confidence': float,
    'artist_terms': str,
    'artist_terms_freq': str,
    'artist_terms_weight': str,
    'year': float
}

# read each CSV file into a separate DataFrame, and concatenate them
cols=list(schema.keys())
temp=[]
for csv_file in glob.glob('*.csv'):
    temp_df = pd.read_csv(csv_file,names=cols)
    temp.append(temp_df)
    
df = pd.concat(temp, ignore_index=True)
```

```python
import numpy as np
pop_avg=np.mean(df['song_hotttnesss'])
df['popularity']= df.song_hotttnesss.map(lambda x: 1 if x > pop_avg else 0)
```

# Feature engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.feature_selection import SelectKBest,f_classif

numeric_features= ['artist_familiarity', 'artist_hotttnesss', 'duration', 'end_of_fade_in', 'key', 
            'key_confidence', 'loudness', 'mode', 'mode_confidence', 'start_of_fade_out', 'tempo', 
            'time_signature', 'time_signature_confidence','year']
text_features = ['title','artist_terms']
label= ['popularity']



# Split data into training and testing sets
df=df[numeric_features+text_features+label]
df.dropna(inplace=True)
df_sample=df[numeric_features+label]


# Tokenize and compute tf-idf on the 'title' column
TF_IDF_NUM_FEATS = 5 
tfidf = TfidfVectorizer(max_features=TF_IDF_NUM_FEATS)
title_tfidf_matrix = tfidf.fit_transform(df['title'])
title_tfidf_df = pd.DataFrame(title_tfidf_matrix.toarray(), columns=["title_tfidf_" + str(i) for i in range(TF_IDF_NUM_FEATS)])

# Transform artist_terms column into an array of strings
df["artist_terms_array"] = df["artist_terms"].apply(lambda x: x.strip('[]').replace("'", "").split(','))

# BoW on artist_terms
BOW_VOCAB_SIZE = 10 
BOW_MIN_DF = 2

# Create a CountVectorizer for 'artist_terms_array' column
cv = CountVectorizer(max_features=BOW_VOCAB_SIZE, min_df=BOW_MIN_DF)

# Fit the CountVectorizer on the dataframe
cv.fit(df["artist_terms_array"].apply(lambda x: ' '.join(x)))

# Transform the dataframe to include the BoW features
artist_terms_bow_matrix = cv.transform(df["artist_terms_array"].apply(lambda x: ' '.join(x)))
artist_terms_bow_df = pd.DataFrame(artist_terms_bow_matrix.toarray(), columns=cv.get_feature_names())

df_final = pd.concat([df_sample, title_tfidf_df, artist_terms_bow_df], axis=1)
df_final.dropna(inplace=True)

# Split data into training and testing sets
X = df_final.drop(label, axis=1)
y = df_final[label]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Preprocess the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform feature selection using SelectKBest and f_classif
selector = SelectKBest(f_classif, k=10)
selector.fit(X_train_scaled, y_train)
X_train_selected = selector.transform(X_train_scaled)
X_test_selected = selector.transform(X_test_scaled)

```

## Training demo

We define a song to be popular if its song_hotttnesss value is above average,

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix

# Instantiate a Random Forest Classifier with 100 estimators
rf = RandomForestClassifier(n_estimators=100,max_depth=15, random_state=42)

# Fit the model to the training data
rf.fit(X_train, y_train)

# Predict the labels of the test data
y_pred = rf.predict(X_test)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Random Forest Classifier Accuracy:", accuracy)

# confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

```

```python
# Fit the model to the training data
rf.fit(X_train_scaled, y_train)

# Predict the labels of the test data
y_pred = rf.predict(X_test_scaled)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Random Forest Classifier Accuracy:", accuracy)

# confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Fit the model to the training data
rf.fit(X_train_selected, y_train)

# Predict the labels of the test data
y_pred = rf.predict(X_test_selected)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Random Forest Classifier Accuracy:", accuracy)

# confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import xgboost as xgb

# Define the XGBoost classifier
xgb_clf = xgb.XGBClassifier()

# Fit the XGBoost classifier to the training data
xgb_clf.fit(X_train_selected, y_train)

# Predict the labels of the test data
y_pred = xgb_clf.predict(X_test_selected)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("XGBoost Classifier Accuracy:", accuracy)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
